# Visualization with Pandas

Visualization is a very important component of any Exploratory Data Analysis. A good Visualization can answer a lot of questions at one.

I am using the below blog to get some onhand understanding of plotting in python - http://tomaugspurger.github.io/modern-6-visualization.html

### 1. Matplotlib
Matplotlib lets you control essentially anything on the figure. An overly-verbose API just means there's an opportunity for a higher-level, domain specific, package to exist (like seaborn for statistical graphics).

### 2. Pandas' builtin-plotting
DataFrame and Series have a .plot namespace, with various chart types available (line, hist, scatter, etc.). 

### 3. Seaborn
Seaborn, created by Michael Waskom, "provides a high-level interface for drawing attractive statistical graphics." Seaborn gives a great API for quickly exploring different visual representations of your data.
