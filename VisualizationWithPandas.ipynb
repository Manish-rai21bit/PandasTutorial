# Visualization with Pandas

Visualization is a very important component of any Exploratory Data Analysis. A good Visualization can answer a lot of questions at one.

I am using the below blog to get some onhand understanding of plotting in python - http://tomaugspurger.github.io/modern-6-visualization.html

### 1. Matplotlib
Matplotlib lets you control essentially anything on the figure. An overly-verbose API just means there's an opportunity for a higher-level, domain specific, package to exist (like seaborn for statistical graphics).

### 2. Pandas' builtin-plotting
DataFrame and Series have a .plot namespace, with various chart types available (line, hist, scatter, etc.). 

### 3. Seaborn
Seaborn, created by Michael Waskom, "provides a high-level interface for drawing attractive statistical graphics." Seaborn gives a great API for quickly exploring different visual representations of your data.

```python
%matplotlib inline 
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)

iris = pd.read_csv('/Users/manishrai/Desktop/DSMLAI/PandasTutorial/Data/IRIS.csv')
iris.head()
```

###### Checking some stats of the dataset

```python
iris.describe() #describe works on the numeric columns
```

```python
iris.info() #info gives the information about all the columns
```

##### Checking the different Species and their count

```python
iris["species"].value_counts()  #filtering out the species column and doing a value count
```

#### We use the pandas .plot 
##### plotting to see the correlation between the length and width of the petel

```python
iris.plot(kind='scatter', x='petal_length', y='petal_width')
```

##### Plotting a relationship between sepal_length and sepal_width 

```python
iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
```

### Now let's try out the Seaborn library

```python
sns.jointplot( x='sepal_length', y='sepal_width', data=iris, size=5, color='red', kind='scatter')
```

```python
sns.jointplot( x='sepal_length', y='sepal_width', data=iris, size=5, color='red', kind='reg')
```

```python
sns.jointplot( x='sepal_length', y='sepal_width', data=iris, size=5, color='red', kind='kde')
```

#### What are the species these points belong to?

```python
sns.FacetGrid(data=iris, hue='species', size = 5).map(plt.scatter, 'sepal_length', 'sepal_width').add_legend()
```

#### Looking at individual feature by species
###### Using Box plot

```python
sns.boxplot(data=iris, x = 'species', y = 'petal_length', width=.5)
```

```python
sns.pairplot(iris, hue="species", size=3)
```
